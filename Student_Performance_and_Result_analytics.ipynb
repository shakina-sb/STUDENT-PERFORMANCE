```python
import pandas as pd
df=pd.read_csv("Student_Performance and Result_analytics.csv")
```

```python
def grade(avg):
    if avg >= 90:
        return "A+"
    elif avg >= 80:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 60:
        return "C"
    else:
        return "D"

df["TOTAL_MARKS"] = df["C++"] + df["Java"] + df["Python"] + df["Data Structure"]
df["Average"] = df["TOTAL_MARKS"] / 3
df["Grade"] = df["Average"].apply(grade)
df["Status"] = df[["C++","Java","Python","Data Structure"]].apply(
    lambda x: "Pass" if all(x >= 40) else "Fail",
    axis=1
)
display(df)
```

```python
subject_columns = ['C++', 'Java', 'Python', 'Data Structure']
average_scores = df[subject_columns].mean()
print('Subject-wise Average Scores:')
print(average_scores.to_string())
```

```python
print("\nBest Performing Subject:", average_scores.idxmax())
print("Lowest Performing Subject:", average_scores.idxmin())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Grade', data=df, order=['A+', 'A', 'B', 'C', 'D'], palette='viridis')
plt.title('Overall Grade Distribution')
plt.xlabel('Grade')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.figure(figsize=(7, 7))
df['Status'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
plt.title('Pass/Fail Distribution')
plt.ylabel('') # Hide the default 'Status' label
plt.show()
```

```python
plot_average_scores_df.columns = ['Subject', 'Average Score']
plt.figure(figsize=(8, 6))
sns.barplot(x='Subject', y='Average Score', data=plot_average_scores_df, palette='coolwarm')
plt.title('Average Score per Subject')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.ylim(0, 100) # Assuming scores are out of 100
plt.show()
```

```python
subject_scores = df[['C++','Java','Python','Data Structure']]
correlation_matrix = subject_scores.corr()

plt.figure(figsize=(7, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='cividis', fmt=".2f")
plt.title('Correlation between Subject Scores')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['C++', 'Java', 'Python','Data Structure']], palette='Pastel1')
plt.title('Distribution of Scores per Subject')
plt.xlabel('Subject')
plt.ylabel('Scores')
plt.ylim(0, 100)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.bar(average_scores.index, average_scores.values)
plt.title("Subject-wise Average Marks")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")
plt.show()
```

```python
class_result = df.groupby("Class").agg(
    Students=("Student Name", "count"),
    Average_Marks=("Average", "mean"),
    Highest_Marks=("Average", "max"),
    Lowest_Marks=("Average", "min"),
    Pass_Count=("Result", lambda x: (x == "Pass").sum()),
    Fail_Count=("Result", lambda x: (x == "Fail").sum())
)
display(class_result)
```
